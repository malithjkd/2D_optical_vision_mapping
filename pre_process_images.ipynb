```python
# import packages
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
image_save_dir = "/home/malithjkd/Documents/cam_test/test9/0.png"
#image = Image.open(image_save_dir)
image1 = cv2.imread(image_save_dir, cv2.IMREAD_GRAYSCALE)
laplacian = cv2.Laplacian(image1,cv2.CV_64F).var()
print(image1.shape, laplacian)
plt.imshow(image1)
```

To undestand images is forcused or not we are using laplacian filter

```python
edges1 = cv2.Canny(image1,45,100)

plt.imshow(edges1,cmap='gray')
plt.savefig('edges1.png')
```

```python
image_save_dir = "/home/malithjkd/Documents/cam_test/test9/4.png"
#image = Image.open(image_save_dir)
image2 = cv2.imread(image_save_dir, cv2.IMREAD_GRAYSCALE)
laplacian = cv2.Laplacian(image2,cv2.CV_64F).var()
print(image2.shape, laplacian)

edges2 = cv2.Canny(image2,45,100)
plt.imshow(edges2,cmap='gray')
plt.savefig('edges2.png')
```

```python
image_red = np.subtract(edges1,edges2)
plt.imshow(image_red,cmap='gray')
```
